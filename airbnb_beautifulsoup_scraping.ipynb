```python
import requests
import urllib.request
import time
import re
from bs4 import BeautifulSoup
```

```python
url = 'http://insideairbnb.com/get-the-data.html'
response = requests.get(url)
response
```

```python
soup = BeautifulSoup(response.text, "html.parser")
soup.find_all('a', string = 'listings.csv')
```

```python
link_with_text = soup.findAll('a', href=re.compile('(2019-12-07)*listings.csv$'))
type(link_with_text)

link = link_with_text[0]['href']
link
```

```python
time.sleep(2)
urllib.request.urlretrieve(link,'./' + link.split('/')[-1]) 
time.sleep(2)
```
